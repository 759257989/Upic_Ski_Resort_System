```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = "/Users/yuwang/Desktop/CS6650/assignment/assignment1/client_part2.csv"  # Update with your actual file path
df = pd.read_csv(file_path)

# Convert timestamps to seconds (assuming relativeStartMillis is in milliseconds)
df['time_seconds'] = df['relativeStartMillis'] // 1000  # Convert to seconds

# Group by second and count requests
throughput = df.groupby('time_seconds').size()

# Plot the throughput over time
plt.figure(figsize=(10, 5))
plt.plot(throughput.index, throughput.values, marker='o', linestyle='-')
plt.xlabel('Time (seconds)')
plt.ylabel('Requests per second')
plt.title('Throughput Over Time')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = "/Users/yuwang/Desktop/CS6650/assignment/assignment1/client_part2.csv"
df = pd.read_csv(file_path)

# Convert relativeStartMillis from milliseconds to seconds and round to nearest second
df['time_seconds'] = (df['relativeStartMillis'] / 1000).astype(int)

# Group by time_seconds and compute average latency
avg_latency = df.groupby('time_seconds')['latencyMillis'].mean()

# Plot average latency over time as a line plot
plt.figure(figsize=(10, 5))
plt.plot(avg_latency.index, avg_latency.values, marker='o', linestyle='-')
plt.xlabel('Time (seconds)')
plt.ylabel('Average Latency (ms)')
plt.title('Average Latency Over Time')
plt.grid(True)
plt.show()
```
